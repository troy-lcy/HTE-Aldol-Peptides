# Sample represent peptide from library and visualize with umap

```python
import numpy as np
import pandas as pd
from utils import kennardstonealgorithm
```

```python
# Using Kennard Stone algorithm to select the most representative points for all data
clustered_df = pd.read_csv('./cluster/results_VPGLA/pca-3-cluster.csv') # cluster results
feature = pd.read_csv('./cluster/results_VPGLA/VPGLA_feature.csv', header=None)
X = feature.values
y = clustered_df.values

selected_sample_numbers = kennardstonealgorithm(X, round(len(feature)*0.05))
```

# Visualize selected samples by UMAP

```python
# visualize data and clusters
import umap
import umap.plot

reducer = umap.UMAP()
# feature_data = X.values
feature_data = X
embedding = reducer.fit_transform(feature_data) # default reduces dimensions to 2d.

mapper = umap.UMAP().fit(feature_data)

umap.plot.output_notebook()

cluster_df = pd.read_csv('./cluster/results_VPGLA/pca-3-cluster.csv')

hover_data = pd.DataFrame({'index':cluster_df.index.tolist(),
                           'label':cluster_df['label'],
                          })

p = umap.plot.interactive(mapper, values=cluster_df['label'],cmap = 'rainbow', hover_data=hover_data, point_size = 5) # theme='darkgreen'

umap.plot.show(p)
```

```python
# visualize data and pointout selectedones
import umap
import umap.plot

reducer = umap.UMAP()
# feature_data = X.values
feature_data = X
embedding = reducer.fit_transform(feature_data) # default reduces dimensions to 2d.

mapper = umap.UMAP().fit(feature_data)

umap.plot.output_notebook()
# selections
cluster_df = pd.read_csv('./cluster/results_VPGLA/pca-3-cluster.csv')
cluster_df.insert(loc= 2, column='value', value= 0)
for i in selected_sample_numbers[0]:
    cluster_df.loc[i,'value'] = 1

hover_data = pd.DataFrame({'index':cluster_df.index.tolist(),
                           'label':cluster_df['label'],
                          })

p = umap.plot.interactive(mapper, values=cluster_df['value'],cmap = 'cool', hover_data=hover_data, point_size = 2) # theme='darkgreen'

umap.plot.show(p)
```
